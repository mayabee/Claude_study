# 텍스트 데이터

- 컴퓨터는 텍스트를 2진수, 10진수등으로 표현함.
- 숫자와 알파벳을 대응하는 방법을 인코딩방식이라 함 ( 다양한 인코딩방식이 있다.)
- ASCII 인코딩방식(한글은 아스키 코드에 없어서 unicode를 사용)
- ASCII : 알파벳, 숫자, 특수문자, 제어문자를 포함한 128개문자를 10진수로 표현
- 인코딩방식: ASCII, UTF-8, UTF-16, EUC-KR, Shift-JIS 등
- 바이너리데이터인코딩 방식: JPEG, PNG, GIF 

* 여기서 압축이라는 것은 이진수를 좀더 적게 표현하는 것이 아닐까? 용량이 줄어들게. 
예)00110001 00110000 00110000 --> 011011 

# 바이너리 데이터

- 텍스트 에디터로 열었을 때 의미를 이해할 수 없는 데이터

- 100 -> 10진수 49 48 48 -> 컴퓨터는 00110001 00110000 00110000 이진수로 읽음
- 만일, 100을 100으로 저장하면, 2진수로는 0110011로 변환되므로 용량절약
- 그러나 ASCII코드표로 텍스트 데이터로 변환하면, 100은 d 로 환원
- 따라서,텍스트 데이터로 변환했을때 알 수 없게 된다.

- 대표적인 바이너리 데이터는 이미지와 동영상 : 텍스트 데이터로 표현할 수 없다. 즉 텍스트 데이터로 읽을 수 없다. 이는 2진수로 변환이 안된다는 뜻

```python
# 딕셔너리로 객체 만들기/여기서는 성적이 아니라 학샘객체로 봄

students=[
    {'name':'윤인성','korean':87,'math':98,"english":88,"science":95},
    {'name':'연하진','korean':84,'math':80,"english":88,"science":95},
    {'name':'구지연','korean':83,'math':70,"english":88,"science":95},
    {'name':'나선주','korean':85,'math':60,"english":88,"science":95},
    {'name':'윤아린','korean':86,'math':92,"english":88,"science":95},
    {'name':'윤명월','korean':87,'math':99,"english":88,"science":95}
]

# 학생을 한명씩 반복하여 출력한다.
import numpy as np
print('이름','총점','평균',sep='\t')
for student in students:
    sum=student['korean']+student['math']+student['english']+student['science']
    average=np.round(sum/4)
    print(f"{student['name']}\t{sum}\t{average}")
```

```python
# 딕셔너리 오타를 방지하고 편리하게 하기 위해, 딕셔너리 생성하는 함수를 만드어보자

def create_student(name, korean,math,english,science):
    students=[]
    student={}
    student['name']=name
    student['koeran']=korean
    student['math']=math
    student['english']=english
    student['science']=science
    students.append(student)
    
create_student('윤인성',87,98,88,95)
create_student('연하진',87,98,88,95)
create_student('구지연',87,98,88,95)
create_student('나선주',87,98,88,95)
create_student('윤아린',87,98,88,95)
create_student('윤명월',87,98,88,95)

for student in students:
    print(student)
   
```
