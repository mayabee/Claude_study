  - Chrome -> V8 엔진                                                                                
  - Firefox -> SpiderMonkey 엔진                                                                     
  - Safari -> JavaScriptCore 엔진                                                                    
                                                                                                     
  그리고 Node.js는 Chrome의 V8 엔진을 브라우저에서 꺼내서 독립적으로 실행할 수 있게 만든 것입니다.   
                                                                                                     
  그래서 같은 자바스크립트 언어인데 실행 환경에 따라 할 수 있는 일이 다릅니다:                       
                  
  ┌────────────────────┬──────────┬─────────┐                                                        
  │                    │ 브라우저 │ Node.js │
  ├────────────────────┼──────────┼─────────┤                                                        
  │ DOM 조작 (화면)    │ O        │ X       │
  ├────────────────────┼──────────┼─────────┤                                                        
  │ 파일 시스템 접근   │ X        │ O       │                                                        
  ├────────────────────┼──────────┼─────────┤                                                        
  │ 네트워크 서버 생성 │ X        │ O       │                                                        
  ├────────────────────┼──────────┼─────────┤                                                        
  │ npm 패키지 사용    │ 제한적   │ O       │
  └────────────────────┴──────────┴─────────┘                                                        
                  
  React가 브라우저에서 화면을 그리는 데 집중하고, Express가 Node.js에서 서버를 처리하는 이유가 이    
  차이 때문입니다

                                                                                                       
❯ 그래서 브라우져처럼 Node.js가 실행환경이므로 Node.js를 설치하면 React로 짠 자바스크립트가          
브라우져는 물론 서버에서도 동작화는 거구나                                                           
                                                                                                     
● 살짝 구분이 필요합니다.                                                                            

  Node.js를 설치하면 자바스크립트 언어 자체는 서버에서도 동작하지만, React 코드가 서버에서 그대로    
  동작하는 건 아닙니다.
                                                                                                     
  React는 DOM(브라우저 화면)을 조작하는 라이브러리라서, 최종 실행은 브라우저에서 됩니다. Node.js는   
  React 개발 과정에서 도우미 역할을 합니다:
                                                                                                     
  [개발 시]                                                                                          
    소스코드(React/JSX)                                                                              
      → Node.js가 빌드/번들링 (Webpack, Babel)                                                       
      → 결과물(HTML, CSS, JS)                                                                        
      → 브라우저에서 실행                                                                            
                                                                                                     
  [서버]                                                                                             
    Node.js + Express → API 서버 (React와는 별개 코드)                                               
                                                                                                     
  정리하면:                                                                                          
  - React 코드 -> Node.js가 빌드해주고, 브라우저에서 실행                                            
  - Express 코드 -> Node.js가 직접 실행하는 서버 코드                                                
  - 둘 다 자바스크립트지만, 실행되는 곳이 다릅니다   
                                                                                                     
  그래서 이번 실습에서 Express 프로젝트를 별도로 초기화하는 것입니다. React(프론트)와                
  Express(백엔드)는 역할이 다른 별개의 프로젝트입니다.  

# Node.js, 자바스크립트, 브라우져, React, Express


❯ React는브라우져에서 동작하는  자바스크립트 라이브러리, Express는 서버에서 동작하는 라이브러리인데, 
 React는 빌드되서 브라우져를 실행환경으로 돌고, Express는 Node.js를 실행환경으로 직접 동작.. 맞아?   
                                                                                                     
● 네, 정확합니다!  

----------------------------------------------------------------------------

❯ 요약하면, Node.js의 결과물(?)이 브라우져에서 실행되고, 서버에서는 Node.js가 직접실행. 이것들을     
개발하는 코드가 node.js, Express 맞나?                                                               
                                                                                                     
● 거의 맞는데, 용어를 조금 정리하면:                                                                 

  Node.js는 개발 도구/언어가 아니라 실행 환경입니다.                                                 
   
  개발하는 코드를 기준으로 보면:                                                                     
                  
  ┌───────────────┬──────────────────────┬────────────────────────┐                                  
  │               │      프론트엔드      │         백엔드         │
  ├───────────────┼──────────────────────┼────────────────────────┤                                  
  │ 작성하는 코드 │ React (자바스크립트) │ Express (자바스크립트) │
  ├───────────────┼──────────────────────┼────────────────────────┤                                  
  │ 실행 환경     │ 브라우저             │ Node.js                │                                  
  ├───────────────┼──────────────────────┼────────────────────────┤                                  
  │ 역할          │ 화면 그리기          │ API, 데이터 처리       │                                  
  └───────────────┴──────────────────────┴────────────────────────┘                                  
                  
  Node.js는 두 곳에서 쓰입니다:                                                                      
  1. 빌드 도구로 - React 코드를 브라우저용으로 변환해주는 역할
  2. 서버 실행 환경으로 - Express 코드를 직접 실행하는 역할                                          
                                                           
  즉, 자바스크립트라는 하나의 언어로 프론트(React)와 백엔드(Express) 모두 개발하고, Node.js가 그     
  중심에서 양쪽을 모두 지원하는 구조입니다.              

# 명령어 해설 : ---- && ----


● cd my-react-app && npm start는 두 개의 명령어를 &&로 연결한 것입니다.                          
                                                                                                   
  ┌─────────────────┬───────────────────────────────────┐                                          
  │      부분       │               의미                │                                          
  ├─────────────────┼───────────────────────────────────┤                                          
  │ cd my-react-app │ my-react-app 폴더로 이동          │
  ├─────────────────┼───────────────────────────────────┤                                          
  │ &&              │ 앞 명령이 성공하면 다음 명령 실행 │                                          
  ├─────────────────┼───────────────────────────────────┤                                          
  │ npm start       │ React 개발 서버 시작              │                                          
  └─────────────────┴───────────────────────────────────┘                                          
                  
  && vs ; 차이:                                                                                    
  - && — 앞 명령이 성공해야 다음 실행 (폴더가 없으면 npm start 안 함)
  - ; — 앞 명령 성공 여부와 관계없이 다음 실행 (폴더가 없어도 npm start 시도 → 오류)               
                                                                                    
  즉 &&를 쓰는 이유는 폴더 이동이 실패했을 때 엉뚱한 위치에서 서버가 실행되는 것을 방지하기        
  위해서입니다.
