# claude code 스터디

쿠버네티스/도커 학습 계획                                           
                                                                      
 Context                                                              
                                                                      
 - 도서: "컨테이너 인프라 환경 구축을 위한 쿠버네티스/도커"           
 - 환경: Docker 컨테이너(Debian/ARM64)                                
 - 제약: VirtualBox/Vagrant 설치 불가 (시스템 복잡화 방지)
 - 목표: 현재 환경에서 최대한 개념 이해 + 가능한 범위만 실습          

 ---                                                                  
 챕터별 학습 전략
                                                                      
 ch2 - 테스트 환경 구성 [이론만]

 - Vagrantfile 코드 읽고 VM 구성 이해 (CPU/RAM/IP 설정 등)
 - 핵심 개념: IaC, 가상화 vs 컨테이너화 차이점 이해
 - 실습 없음

 ch3 - 쿠버네티스 기본 [이론 + 선택적 실습]

 - YAML 파일 분석 (Pod, Deployment, Service, PV/PVC, StatefulSet 등)
 - 핵심 개념: 쿠버네티스 아키텍처, 오브젝트 종류와 역할
 - 선택적 실습: minikube 또는 k3d로 로컬 클러스터 구성 가능
   - Docker만으로 동작하므로 시스템 부담 최소화

 ch4 - Docker 심화 [실습 가능]

 - Dockerfile 분석 (기본/최적화/멀티스테이지 빌드)
 - 핵심 개념: 이미지 레이어, 빌드 최적화, 사설 레지스트리
 - 실습: Dockerfile 직접 작성/빌드 (현재 환경에서 가능)

 ch5 - Jenkins CI/CD [이론 + 선택적 실습]

 - Jenkinsfile 분석 (선언적 파이프라인 문법)
 - 핵심 개념: CI/CD 파이프라인, GitOps, Kustomize/Helm 활용
 - 선택적 실습: k3d 환경에서 Jenkins 배포 가능

 ch6 - Prometheus & Grafana [이론 + 선택적 실습]

 - values.yaml 분석 (Helm 차트 설정)
 - 핵심 개념: 메트릭 수집, 시각화, AlertManager 알림 설정
 - 선택적 실습: Docker Compose로 모니터링 스택 구성 가능

 app - 부록 [이론만]

 - kubectl 자동완성, kubespray 개념, 컨테이너 네임스페이스 이론 이해

 ---
 학습 우선순위

 1. ch4 Docker 심화 (현재 환경에서 바로 실습 가능 - 최우선)
 2. ch3 쿠버네티스 기본 (YAML 분석 + k3d 선택 실습)
 3. ch5 Jenkins CI/CD (개념 이해 + 파이프라인 분석)
 4. ch6 모니터링 (개념 이해)
 5. ch2 환경 구성 (이론만, 마지막)
